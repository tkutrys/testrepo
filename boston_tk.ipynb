####  Question 1: Is there a significant difference in the Median value of houses bounded by the Charles River or not?Boxplot viIs there a significant difference in the Median value of houses bounded by the Charles River or not?

Hypothesis:
•	Null Hypothesis (H0): There is no significant difference in the median value of houses bounded by the Charles River.
•	Alternative Hypothesis (H1): There is a significant difference in the median value of houses bounded by the Charles River.
2.	Significance Level (α): α = 0.05 (This indicates that we are willing to accept a 5% chance of rejecting the null hypothesis when it is true, known as a Type I error.)
3.	Test Statistics:
•	Since we are comparing the median values of two independent samples (houses bounded by the Charles River and houses not bounded by the Charles River), a suitable statistical test is the independent samples t-test.
•	We will calculate the test statistic and the corresponding p-value using the t-test.
4.	Conclusion from the test:
•	If the p-value is less than the significance level (α), we reject the null hypothesis. Otherwise, we fail to reject the null hypothesis.

###  A Boxplot visualization for the Median value of Owner-Occupied Homes

####  Question 2: Is there a difference in Median value of houses (MEDV) for each porportion of owner occupied units built prior to 1940 (Age)? ANOVA

1.	Hypothesis:
•	Null Hypothesis (H0): There is no significant difference in the median values of houses for each proportion of owner-occupied units built prior to 1940.
•	Alternative Hypothesis (H1): There is a significant difference in the median values of houses for each proportion of owner-occupied units built prior to 1940.
2.	Significance Level (α): α = 0.05 (This indicates that we are willing to accept a 5% chance of rejecting the null hypothesis when it is true, known as a Type I error.)
3.	Test Statistics:
•	Since we are comparing the median values of multiple groups (proportions of owner-occupied units built prior to 1940), a suitable statistical test is the Analysis of Variance (ANOVA) test.
•	We will calculate the test statistic and the corresponding p-value using the ANOVA test.
4.	Conclusion from the test:
•	If the p-value is less than the significance level (α), we reject the null hypothesis. This indicates that there is a significant difference in the median values of houses across the different proportions of owner-occupied units built prior to 1940.
•	If the p-value is greater than the significance level (α), we fail to reject the null hypothesis. This suggests that there is not enough evidence to conclude that there is a significant difference in the median values of houses across the different proportions of owner-occupied units built prior to 1940.

###  A Boxplot for the MEDV variable vs the Age variable (discretizing the age variable into 3 groups of 35 years and younger, between 35 and 70 years old, and 70 years and older).

```python
# Discretize the 'AGE' variable into three groups
bins = [0, 35, 70, float('inf')]
labels = ['35 years and younger', 'between 35 and 70 years', '70 years and older']
boston_df['AGE_group'] = pd.cut(boston_df['AGE'], bins=bins, labels=labels)

# Create a boxplot
plt.figure(figsize=(8, 6))
boston_df.boxplot(column='MEDV', by='AGE_group')

# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('Median value of owner-occupied homes ($1000\'s)')
plt.title('Boxplot of Median Home Values by Age Group')

# Display the plot
plt.show()
```

####  Question 3: Can we conclude that there is no relationship between Nitric Oxide concentrations and proportions of non-retail business acres per town?

1.	Hypothesis:
•	Null Hypothesis (H0): There is no significant relationship between nitric oxide concentrations and the proportion of non-retail business acres per town.
•	Alternative Hypothesis (H1): There is a significant relationship between nitric oxide concentrations and the proportion of non-retail business acres per town.
2.	Significance Level (α): α = 0.05 (This indicates that we are willing to accept a 5% chance of rejecting the null hypothesis when it is true, known as a Type I error.)
3.	Test Statistics:
•	We will calculate the Pearson correlation coefficient (r) between the NOX and INDUS variables. The Pearson correlation measures the linear relationship between two variables and ranges from -1 to +1. A value close to +1 indicates a strong positive relationship, a value close to -1 indicates a strong negative relationship, and a value close to 0 indicates no linear relationship.
4.	Conclusion from the test:
•	If the p-value associated with the Pearson correlation coefficient is less than the significance level (α), we reject the null hypothesis and conclude that there is a significant relationship between nitric oxide concentrations and the proportion of non-retail business acres per town.
•	If the p-value is greater than the significance level (α), we fail to reject the null hypothesis and conclude that there is not enough evidence to support a significant relationship between the variables
